## Tic-Tac-Toe Game by Sanjay Panchal

```python
def print_board(board):
    print(board[0]+"|"+board[1]+"|"+board[2])
    print("-+-+-")
    print(board[3]+"|"+board[4]+"|"+board[5])
    print("-+-+-")
    print(board[6]+"|"+board[7]+"|"+board[8])
    
def play_game():
    print("*"*41)
    print("   Welcome to the Tic-Tac-Toe Game !!!")
    print("*"*41)
    print("")
    board = [" "," "," "," "," "," "," "," "," "]
    print_board(board)
    player = ["X", "O"]
    current = player[0]
    game_over=False
    
    while not game_over:
        print_board(board)
        print("\nIts ",current," Turn\n")
        n = int(input("Enter Your Input (1-9) : "))-1
        print("")
        
        if board[n] == " ":
            board[n] = current
        else:
            print("Please Enter a Valid Move !!!")
            
        if (board[0]==board[1]==board[2]!=" " or
            board[3]==board[4]==board[5]!=" " or
            board[6]==board[7]==board[8]!=" " or
            board[0]==board[3]==board[6]!=" " or
            board[1]==board[4]==board[7]!=" " or
            board[2]==board[5]==board[8]!=" " or
            board[0]==board[4]==board[8]!=" " or
            board[2]==board[4]==board[6]!=" "):
            print_board(board)
            print("")
            print("-"*10)
            print(current, " Win's")
            print("-"*10)
            game_over=True
            break
            
        if " " not in board:
            print_board(board)
            print("")
            print("-"*12)
            print("Its Tie !!!")
            print("-"*12)
            game_over=True
            break
            
        if current == player[0]:
            current=player[1]
        else:
            current=player[0]
        
play_game()
```
